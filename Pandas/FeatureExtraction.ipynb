```python
## extracting the columns of a data episodes to count the no of the epsiodes
def extract_ep(txt):
    data = ""
    check = False
    
    for i in txt:
        if i == ')':
            break
        if i == '(':
            check = True
        if check == True:
            data += i
    return data 
```

```python
df['Ep']=df['Title'].apply(extract_ep)
```

```python
def extracting_timestamp(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
            for j in range(i, i+18):
                data += txt[j] 
            return data
```

```python
def extracting_timestamp(txt):
    for i in range(len(txt)):
        if txt[i] == ')':
            return txt[i+1:i+18]
```

```python
df['Total Time']=df['Title'].apply(extracting_timestamp)
```
